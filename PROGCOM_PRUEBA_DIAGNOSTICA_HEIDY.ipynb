```python
import math

# Calcular área de un triángulo
base = float(input("Ingrese la base del triángulo: "))
altura = float(input("Ingrese la altura del triángulo: "))
print(f"El área del triángulo es: {base * altura / 2}")

# Convertir grados Celsius a Fahrenheit
num = float(input("Ingrese la temperatura en grados Celsius: "))
fahrenheit = num * 9/5 + 32
print(f"La temperatura en grados Fahrenheit es: {fahrenheit}")

# Calcular el promedio de tres números
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
num3 = float(input("Ingrese el tercer número: "))
print(f"El promedio de los números es: {(num1 + num2 + num3) / 3}")

# Calcular perímetro y área de un círculo
rad = float(input("Ingrese el radio del círculo: "))
print(f"El perímetro del círculo es: {2 * math.pi * rad}")
print(f"El área del círculo es: {math.pi * rad**2}")

# Resolver ecuación cuadrática ax² + bx + c = 0
a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))
c = float(input("Ingrese el valor de c: "))

discriminante = b**2 - 4*a*c
if discriminante < 0:
    print("La ecuación no tiene soluciones reales.")
else:
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)
    print(f"Las soluciones de la ecuación son: {x1} y {x2}")

# Determinar si un número es par o impar
num = int(input("Ingrese un número: "))
if num % 2 == 0:
    print(f"{num} es un número par")
else:
    print(f"{num} es un número impar")

# Determinar si un año es bisiesto
año = int(input("Ingrese el año: "))
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    print(f"{año} es un año bisiesto")
else:
    print(f"{año} no es un año bisiesto")

# Determinar si una persona pertenece a la tercera edad
nam = input("Ingrese su nombre: ")
e = int(input("Ingrese su edad: "))
if e >= 70:
    print(f"{nam}, se encuentra en el rango de preferencia en fila, continúe")
else:
    print(f"{nam}, no se encuentra en el rango de preferencia en la fila. Continúe en espera")

# Comparar tres números y mostrar el mayor
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
num3 = float(input("Ingrese el tercer número: "))

if num1 > num2 and num1 > num3:
    print(f"El número mayor es: {num1}")
elif num2 > num1 and num2 > num3:
    print(f"El número mayor es: {num2}")
else:
    print(f"El número mayor es: {num3}")

# Determinar el tipo de triángulo según sus lados
la1 = float(input("Ingrese el primer lado del triángulo: "))
la2 = float(input("Ingrese el segundo lado del triángulo: "))
la3 = float(input("Ingrese el tercer lado del triángulo: "))

if la1 == la2 == la3:
    print("El triángulo es equilátero")
elif la1 == la2 or la2 == la3 or la1 == la3:
    print("El triángulo es isósceles")
else:
    print("El triángulo es escaleno")
```
