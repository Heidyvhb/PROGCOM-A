```python
import json
from abc import ABC, abstractmethod
from datetime import datetime

# Abstracción
class Juego(ABC):
    @abstractmethod
    def cargar_preguntas(self):
        pass

    @abstractmethod
    def cargar_respuestas(self):
        pass

    @abstractmethod
    def jugar(self):
        pass

# Herencia
class JuegoPreguntas(Juego):
    def init(self):
        self.preguntas = []
        self.respuestas = {}
        self.resultado = 0

    def cargar_preguntas(self):
        with open("preguntas.json", encoding="utf-8") as f:
            self.preguntas = json.load(f)

    def cargar_respuestas(self):
        with open("respuestas.json", encoding="utf-8") as f:
            self.respuestas = json.load(f)

    def jugar(self):
        nombre = input("👤 Ingresa tu nombre: ")
        print(f"\n🎮 ¡Bienvenido/a {nombre}! Comencemos el juego:\n")
        puntos = 0

        for pregunta in self.preguntas:
            print(pregunta["texto"])
            for clave, opcion in pregunta["opciones"].items():
                print(f"  {clave}) {opcion}")
            respuesta_usuario = input("👉 Tu respuesta: ").lower()

            if respuesta_usuario == self.respuestas[str(pregunta["id"])]:
                print("✅ ¡Correcto!\n")
                puntos += 1
            else:
                print("❌ Incorrecto.\n")

        self.resultado = puntos
        hora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open("resultados.txt", "a", encoding="utf-8") as f:
            f.write(f"{hora} | {nombre} obtuvo {puntos}/10\n")

        print(f"\n🎉 Fin del juego, {nombre}!")
        print(f"📝 Tu puntaje final fue: {puntos}/10")

# Ejecutar el juego
juego = JuegoPreguntas()
juego.cargar_preguntas()
juego.cargar_respuestas()
juego.jugar()
```
