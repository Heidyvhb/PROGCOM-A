##Estructura PARA - FOR
* Se debe repetir una operaci칩n
* Se conoce cu치ntas veces se debe repetir la operaci칩n

Forma general
```

for (variable) in range (valor inicial,valor final, paso):
  Aqui se coloca con todo lo que se va a repetir n veces

```

Forma especifica (n칰mero de veces)

```

for (variable) in range (n):
  Aqui se coloca todo lo que se va a repetir n veces

```

Trabajando con listas-diccionarios

```

  Aqu칤 se coloca todo lo que se va a repetir n veces

##Estructura de repeticion MIENTRAS - WHILE
Se emplea cuando **NO** se sabe cuantas veces se debe repetir una operaci칩n
1. Una condicion inicial
2. Una condicion final
3. un mecanismo que altere la condicion inicial



```
ci=condicion inicial
while (cf)
  operaciones...
  mecanismo de alteracion de ci
```
A veces, la condicion finl se remplaza por true
Se usan banderas que alteran de true(verdadero) a false (falso).

```
while true:
  el modificador de condicion inicial debe dar
  false
```




````python
opc=input("En que color estoy pensando?")#condicion inicial
while opc!=color_cpu:#condicion final
  print("Perdiste")
  opc=input("En que color estoy pensando?")#Modificador
print("Ganaste")
````

````python
# 26. Tablas de multiplicar del 1 al 10
print("Usando for:")
for i in range(1, 11):
    print(f"Tabla del {i}:")
    for j in range(1, 11):
        print(f"{i} x {j} = {i * j}")
    print()
````

````python
# 26. Tablas de multiplicar del 1 al 10
print("Usando while:")
i = 1
while i <= 10:
    print(f"Tabla del {i}:")
    j = 1
    while j <= 10:
        print(f"{i} x {j} = {i * j}")
        j += 1
    i += 1
    print()
````

````python
# 27. N칰meros perfectos hasta n
def es_numero_perfecto(num):
    suma = sum(i for i in range(1, num) if num % i == 0)
    return suma == num

def encontrar_numeros_perfectos_for(n):
    for num in range(1, n + 1):
        if es_numero_perfecto(num):
            print(num)

def encontrar_numeros_perfectos_while(n):
    num = 1
    while num <= n:
        if es_numero_perfecto(num):
            print(num)
        num += 1

n = int(input("Ingresa un n칰mero l칤mite para los n칰meros perfectos: "))
print("Usando for:")
encontrar_numeros_perfectos_for(n)
print("Usando while:")
encontrar_numeros_perfectos_while(n)
````

````python
# 28. Patr칩n de diamante con asteriscos
def imprimir_diamante_for(n):
    for i in range(1, n + 1, 2):
        print(" " * ((n - i) // 2) + "*" * i)
    for i in range(n - 2, 0, -2):
        print(" " * ((n - i) // 2) + "*" * i)

def imprimir_diamante_while(n):
    i = 1
    while i <= n:
        print(" " * ((n - i) // 2) + "*" * i)
        i += 2
    i -= 4
    while i > 0:
        print(" " * ((n - i) // 2) + "*" * i)
        i -= 2

size = int(input("Ingresa un n칰mero impar para el tama침o del diamante: "))
print("Usando for:")
imprimir_diamante_for(size)
print("Usando while:")
imprimir_diamante_while(size)
````

````python
# 29. N칰meros Armstrong hasta n
def es_armstrong(num):
    digitos = [int(d) for d in str(num)]
    suma = sum(d ** len(digitos) for d in digitos)
    return suma == num

def encontrar_armstrong_for(n):
    for num in range(n + 1):
        if es_armstrong(num):
            print(num)

def encontrar_armstrong_while(n):
    num = 0
    while num <= n:
        if es_armstrong(num):
            print(num)
        num += 1

n = int(input("Ingresa un n칰mero l칤mite para los n칰meros Armstrong: "))
print("Usando for:")
encontrar_armstrong_for(n)
print("Usando while:")
encontrar_armstrong_while(n)
````

````python
# 30. Simulaci칩n de reloj digital
def reloj_digital_for(horas):
    for h in range(horas):
        for m in range(60):
            for s in range(60):
                print(f"{h:02}:{m:02}:{s:02}")

def reloj_digital_while(horas):
    h, m, s = 0, 0, 0
    while h < horas:
        while m < 60:
            while s < 60:
                print(f"{h:02}:{m:02}:{s:02}")
                s += 1
            s = 0
            m += 1
        m = 0
        h += 1

horas = int(input("Ingresa el n칰mero de horas para simular el reloj: "))
print("Usando for:")
reloj_digital_for(horas)
print("Usando while:")
reloj_digital_while(horas)
````
